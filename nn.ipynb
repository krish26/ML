```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#random numbers 

#generator  _ generates numbers 
#Discriminator -job is to stop numbers if not real


def build_discriminator(optimizer = 'adam' , loss = 'binary_crossentropy', metrics=['accuracy']):
    model = Sequential()

    #input layer 
    model.add(Dense(128,activation='relu',input_shape=(1,))) #128 circles in next layer, and 1 in input layer

    #hidden layers 

    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))


    #output layer 
    model.add(Dense(1, activation='sigmoid'))


    #relu is what activates or deactivates neuron , the neuron will gets an value and relu checks value 
    #if number is negative its converted to 0 , and if its 0 its not activated 

    #sigmod - 

    model.compile(optimizer=optimizer , loss=loss , metrics = metrics)
    return model
```
