```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score,accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv('Data/diabetes.csv')
print(len(dataset))
dataset.head()
```

```python
#remove  zeros 

Zero_not_accepted = ['Glucose', 'BloodPressure' , 'SkinThickness' , 'BMI' , 'Insulin']

for column in Zero_not_accepted:
    dataset[column] = dataset[column].replace(0,np.nan)   #replacing o with NAn 
    mean = int(dataset[column].mean(skipna=True)) #cal the mean of na values
    dataset[column] = dataset[column].replace(np.nan,mean) # replacing Nan with mean
```

```python
print(dataset['Glucose'].notnull().sum())
```

```python
#split the data set 
#in the data column 9 is outcome i.e the answer so we dont use it in training 

X = dataset.iloc[: , 0:8]  # iloc is pandas property used to access data with indices (: is all rows )
y = dataset.iloc[: , 8]

X_train, X_test, y_train,y_test = train_test_split(X, y, random_state=0, test_size=0.2)

```

```python
#scaling the data is helpful 
#only scale the data going in i.e X not y
'''
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

'''
#modern way using pipelines 


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])

pipeline.fit(X_train, y_train)

```

```python
import math

math.sqrt(len(y_test))
```

```python
#defining Model 

classifier = KNeighborsClassifier(n_neighbors=11, p=2,  metric = 'euclidean')
```

```python
#fit model 


classifier.fit(X_train, y_train)
```

```python
# Predict the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Evaluate Model
cm = confusion_matrix(y_test, y_pred)
print (cm)
print(f1_score(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
