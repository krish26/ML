```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./data/house_price.csv')  
```

```python
df.describe() #gives statistical summary
```

```python
df_filtered =df['sq_meter']>82
df[df_filtered] #boolean masking
```

```python
df.head()
```

```python
df_filtered=df[df['bedrooms']>2]
df_filtered
```

```python
df_sorted=df.sort_values(by='sales_price',ascending=True) #can also write just 'sales_price'
df_sorted
```

```python
sum= df['sales_price'].sum()
print(sum)
avg= df['sales_price'].mean() #average
print (avg)
```

```python
#for price per sq meter divide sales_price by sq_meter

df["price_per_sqmtr"] = df['sales_price'] / df['sq_meter']
df.head()
```

```python
df_grouped=df.groupby('neighborhood')['sales_price'].mean().reset_index()
df_grouped
```

```python
df_grouped.plot(kind='bar', x='neighborhood', y='sales_price' )
plt.title('Average Sales Price by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average Sales Price')
plt.show()
```
